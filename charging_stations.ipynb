```python
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('RS_Session_256_AU_2673_1.csv')
df.head()
```

```python
df.info()
```

```python
df = df.rename(columns={'Charging Stations': 'charging_stations'})
df = df.rename(columns={'City/Highway': 'location'})
df.head()
```

```python
#for cities
location_list= []	
cs_list = []
for index, row in df.iterrows():
    if(row['Category'] == 'City'):
        location_list.append(row['location'])
        cs_list.append(row['charging_stations'])
print(location_list)
print(cs_list)
```

```python
def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{:.1f}%\n({v:d})'.format(pct, v=val)
    return my_format
fig = plt.figure(figsize =(10, 7))
plt.pie(cs_list, labels = location_list,autopct=autopct_format(cs_list))
 
# show plot
plt.show()
```

```python
#WITH THE GIVEN DATA we can conclude that
""" In cities category Delhi has the highest number
of charging stations with having 94 charging stations 
occupying 23.9% of the total charging stations and Lucknow
with only 1 charging stations occupying 0.3%"""
```
